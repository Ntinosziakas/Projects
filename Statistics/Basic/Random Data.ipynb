```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd

filedata = np.genfromtxt('Random Data.txt', delimiter=',')
filedata = filedata.astype('float32')
#print(np.round((filedata),4))
print("----------------------------------------------------------------------------")


mean=np.round(np.average(filedata),2)
var=np.round(np.var(filedata),4)
sd=np.round(np.std(filedata),4)
max=np.round(np.max(filedata),2)
min=np.round(np.min(filedata),2)
range = max - min
median = np.round(np.median(filedata),2)

print("Descriptive Statistics")
print("----------------------------------------------------------------------------")
print("Mean = ", mean)
print("Variance = ", var)
print("Standard Deviation = ", sd)
print("Max = ", max)
print("Min = ", min)
print("Range = ", range)
print("Median =", median)

filedata[filedata > np.average(filedata)]
sum=np.count_nonzero(filedata)
ba=np.count_nonzero(filedata > np.average(filedata))
tba=ba/sum

print("The number of data which was more than the average :"
      ,np.count_nonzero(filedata > np.average(filedata)),
      "=",np.round(tba,4),"=",np.round(100*tba,2),"%")
print("The number of data which was less than the average :"
      ,np.count_nonzero(filedata < np.average(filedata)),
      "=",np.round(1-tba,4),"=",np.round(100*(1-tba),2),"%")
print("----------------------------------------------------------------------------")

# Sort Data
sfiledata=np.sort(filedata)

# Plot Data
plt.plot(filedata)
plt.title("Line Plot") 
plt.show()

# Histogram
plt.hist(sfiledata)
plt.title("Histogram")
plt.show()

# Boxplot
plt.boxplot(sfiledata)
plt.title("Boxplot") 
plt.show()

plt.bar(filedata, height=1+max, width=0.0088)
plt.title("Bar Chart") 
plt.show()



```
